```python
import os
import cv2
import numpy as np
import requests
import base64
import time
from ultralytics import YOLO
from PIL import Image
from io import BytesIO
import torch
import seaborn as sns
import matplotlib.pyplot as plt
from ultralytics import YOLO
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score,accuracy_score
from torchmetrics.detection import MeanAveragePrecision
```

```python
# Load the custom LeYOLO model
model = YOLO("yolov8n.pt")

# Get class labels from the model
class_names = model.names  # Dictionary mapping class index to class name

def detect_from_url(image_input):
    try:
        if image_input.startswith("data:image/"):
            # Handle base64 string
            header, encoded = image_input.split(",", 1)
            image_data = base64.b64decode(encoded)
            image = Image.open(BytesIO(image_data))
            image = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

        elif image_input.startswith("http://") or image_input.startswith("https://"):
            # Handle standard image URL
            response = requests.get(image_input)
            if response.status_code == 200:
                image = Image.open(BytesIO(response.content))
                image = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
            else:
                print("❌ Failed to download image from URL.")
                return
        else:
            print("⚠️ Invalid input. Provide a valid image URL or base64 string.")
            return

        detect_and_display(image, "LeYOLO Object Detection")

    except Exception as e:
        print(f"🚨 Error during detection: {e}")

# Function to detect objects from a local static image
def detect_from_static_image(image_path):
    image = cv2.imread(image_path)
    if image is None:
        print("Failed to load image.")
        return

    detect_and_display(image, "LeYOLO Object Detection (Static Image)")

# Function to detect objects from a live webcam
def detect_from_webcam():
    cap = cv2.VideoCapture(0)  # Open webcam

    # Set lower resolution for faster processing
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

    prev_time = 0  # For FPS calculation

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame.")
            break

        # FPS Calculation
        curr_time = time.time()
        fps = 1 / (curr_time - prev_time)
        prev_time = curr_time

        # Reduce frame size (optional for performance)
        frame_resized = cv2.resize(frame, (640, 480))

        # Run inference
        results = model.predict(frame_resized, verbose=False)  # Faster execution

        # Draw bounding boxes and labels
        for result in results:
            for box in result.boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates
                conf = box.conf[0].item()  # Confidence score
                cls = int(box.cls[0].item())  # Class ID
                label = f"{class_names.get(cls, 'Unknown')} {conf:.2f}"

                # Draw bounding box
                cv2.rectangle(frame_resized, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame_resized, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Show FPS on the frame
        cv2.putText(frame_resized, f"FPS: {int(fps)}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

        # Display the frame
        cv2.imshow("LeYOLO Object Detection (Live Webcam)", frame_resized)

        # Press 'q' to quit
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

    cap.release()
    cv2.destroyAllWindows()

# Function to perform object detection and display results
def detect_and_display(image, window_name):
    results = model(image)

    # Draw bounding boxes and labels
    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates
            conf = box.conf[0].item()  # Confidence score
            cls = int(box.cls[0].item())  # Class ID
            label = f"{class_names.get(cls, 'Unknown')} {conf:.2f}"

            # Draw bounding box
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Show the image
    cv2.imshow(window_name, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Ask user for input **before** calling any function
choice = input("Enter 'url' for image URL, 'webcam' for live detection, or 'image' for a static image file: ").strip().lower()

if choice == "url":
    image_url = input("Enter image URL: ").strip()
    detect_from_url(image_url)
elif choice == "webcam":
    detect_from_webcam()
elif choice == "image":
    image_path = input("Enter image file path: ").strip()
    detect_from_static_image(image_path)
else:
    print("Invalid choice. Please enter 'url', 'webcam', or 'image'.")
```
